```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df_coronga = pd.read_csv('covid25042020.csv',sep=';',encoding='ISO-8859-1')
```

```python
df_coronga.head()
```

### Sem Subplots

```python

dados_MG = df_coronga[df_coronga['estado'] == 'MG']
dados_SP = df_coronga[df_coronga['estado'] == 'SP']
dados_RJ = df_coronga[df_coronga['estado'] == 'RJ']
dados_ES = df_coronga[df_coronga['estado'] == 'ES']
for i in range(len(dados_SP.data)):
  dados_SP.data.values[i] = i
for i in range(len(dados_MG.data)):
  dados_MG.data.values[i] = i
for i in range(len(dados_RJ.data)):
  dados_RJ.data.values[i] = i 
for i in range(len(dados_ES.data)):
  dados_ES.data.values[i] = i 
plt.figure(figsize=(20,8))
plt.plot(dados_MG['data'],dados_MG['obitosAcumulados'],color = 'green',label = 'MG')
plt.plot(dados_SP['data'],dados_SP['obitosAcumulados'],color = 'red',label='SP')
plt.plot(dados_RJ['data'],dados_RJ['obitosAcumulados'],color = 'yellow',label='RJ')
plt.plot(dados_ES['data'],dados_ES['obitosAcumulados'],color = 'magenta',label='ES')
plt.xlabel('Dias')
plt.ylabel('Obitos')
plt.title('Obitos na regiÃ£o Sudeste')
plt.grid()
plt.legend()
plt.show()
```

### Plotando com o pandas 

```python
estados_norte = ['AM','AC','RO','PA','TO','AP','RR']
estados_nordeste = ['MA','PI','CE','RN','PB','PE','AL','SE','BA']
estados_centro_oeste = ['MT','GO','MS']
estados_sudeste = ['MG','SP','RJ','ES'] 
estados_sul = ['PR','SC','RS']
```

```python
def plota_regiao(dataframe,regiao,axis,titulo = 'TÃ­tulo'):
    '''This function plot dates for brazillian region'''
    for i in regiao:
        dataframe[dataframe['estado'] == i].plot(x = 'data',y = 'obitosAcumulados',ax = axis,label = i,grid =True)
    axis.set(xlabel = 'Datas',ylabel = 'Obitos',title = titulo)
```

```python
fig2,axis = plt.subplots(3,2,figsize = (20,15))
fig2.subplots_adjust(hspace = 0.3)
plota_regiao(df_coronga,estados_sudeste,axis[0,0],'Sudeste')
plota_regiao(df_coronga,estados_nordeste,axis[0,1],'Nordeste')
plota_regiao(df_coronga,estados_norte,axis[1,0],'Norte')
plota_regiao(df_coronga,estados_sul,axis[1,1],'Sul')
plota_regiao(df_coronga,estados_centro_oeste,axis[2,0],'Centro Oeste')
df_coronga.filter(items = ['regiao','obitosNovos']).groupby('regiao').sum().plot.bar(ax = axis[2,1],grid = True)
axis[2,1].set(title = 'Regioes',xlabel = 'Regioes',ylabel = 'Obitos')
fig2.suptitle('NÃºmero de Ã³bitos por regiÃ£o',fontsize = 20)
```
